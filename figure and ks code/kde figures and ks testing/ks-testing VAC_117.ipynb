```python
df= pd.read_csv("/Users/caranix/Library/CloudStorage/OneDrive-TheOhioStateUniversity/CS MS/speeches_VAC_117th/speeches_VACES_117.csv")

```

```python
df["party_2"]= np.where(df["party"] == "R", "R", "D")
```
