```python
#extract time in video
def extract_time(soup_tr_index): 
    time= re.search(r'\d{2}:\d{2}:\d{2}', str(soup_tr_index)).group(0)
    return(time)

#extract name and party-state 
def extract_name_partystate(soup_tr_index):
    name= re.split("<strong>|</strong>", str(soup_tr_index))[1].split(",")[0]
    if re.split("<strong>|</strong>", str(soup_tr_index))[1] == "" or "," not in re.split("<strong>|</strong>", str(soup_tr_index))[1]: 
        party_state = "" 
    else: 
        party_state= re.split("<strong>|</strong>", str(soup_tr_index))[1].split(",")[1].strip()
    
    return(name,party_state)
```

```python
# extract the text! 
def extract_text(tr_list, index): 
    text_old= re.sub('(?<![\r\n])(\r?\n|\n?\r)(?![\r\n])', '', tr_list[index])
    list_full= text_old.lower().split("\n")
    list_partial= [i for i in list_full if i != ""]
    text= list_partial[-1]
    return (text) 
```

```python
# preininitalize columns of df 
colnames= ["i","url", "time", "name", "party-state", "text", 'congress',]
df = pd.DataFrame(columns = colnames)
```

```python
# for each url/video... 
for i in range(0, len(link_to_vid_list)): 
    print(i)
    url = link_to_vid_list[i]
    driver.get(url)
    web_text= driver.page_source
    soup = BeautifulSoup(web_text, 'html.parser')
    soup_tr= soup.find_all('tr')

    # alt tr - used for extracting text 
    tr_list = [] 
    for tr in soup.find_all('tr'):
        tr_list.append(tr.text) 
        
    for index in range(0,len(soup_tr)): 
        time_i= extract_time(soup_tr[index])
        name_i, party_i = extract_name_partystate(soup_tr[index])
        text_i = extract_text(tr_list, index)
        

        
        list_i = [i, url, time_i, name_i, party_i, text_i, "116"]
        df.loc[len(df)] = list_i
    
    
    
    
    
    
    
    
```
